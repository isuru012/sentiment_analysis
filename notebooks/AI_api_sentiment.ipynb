```python
import google.generativeai as genai
import os

# 1. Configuration
# Replace 'YOUR_API_KEY' with your actual key
os.environ["GEMINI_API_KEY"] = "AIzaSyAH7ptCez20WCEh7VnSjP8HO3tPHrCNOWI"
genai.configure(api_key=os.environ["GEMINI_API_KEY"])

# 2. Initialize the Model
# 'gemini-1.5-flash' is faster and cheaper, perfect for simple tasks like sentiment
model = genai.GenerativeModel('gemini-2.5-flash')

def analyze_sentiment(text):
    # 3. Construct the Prompt
    # We explicitly tell the model how to format the output
    prompt = f"""
    Analyze the sentiment of the following text. 
    Classify it as either: Positive, Negative, or Neutral.
    Provide a short explanation for your classification.
    
    Text: "{text}"
    """
    
    try:
        response = model.generate_content(prompt)
        return response.text
    except Exception as e:
        return f"Error: {e}"

# 4. Test with examples
comments = [
    "I absolutely love this product! It saved me so much time.",
    "The service was terrible and the app keeps crashing.",
    "It's okay, does what it says but nothing special."
]

print("--- Sentiment Analysis Results ---")
for comment in comments:
    result = analyze_sentiment(comment)
    print(f"\nInput: {comment}")
    print(f"Analysis:\n{result}")
    print("-" * 30)
```
